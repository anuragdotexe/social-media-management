```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.ensemble import IsolationForest

# Set up visualization style
sns.set(style="whitegrid")

# Load your data
# For this example, we'll generate synthetic data
# Replace this with your data loading code
np.random.seed(42)
data = pd.DataFrame({
    'app': ['Social Media', 'Productivity', 'Entertainment', 'Education', 'Others']*20,
    'usage_time': np.random.randint(0, 120, size=100)  # usage time in minutes
})

# Data exploration
print(data.head())

# Visualize usage time by app category
plt.figure(figsize=(10, 6))
sns.boxplot(x='app', y='usage_time', data=data)
plt.title('Usage Time by App Category')
plt.xticks(rotation=45)
plt.show()

# Feature engineering
data_pivot = data.pivot_table(index=data.index, columns='app', values='usage_time', fill_value=0)

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_pivot)

# Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(data_scaled)
data_pivot['cluster'] = clusters

# Visualize clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_pivot['Social Media'], y=data_pivot['Productivity'], hue=data_pivot['cluster'], palette='Set2')
plt.title('Cluster Analysis of App Usage')
plt.xlabel('Social Media Usage (minutes)')
plt.ylabel('Productivity Usage (minutes)')
plt.legend(title='Cluster')
plt.show()

# Anomaly Detection
iso_forest = IsolationForest(contamination=0.1, random_state=42)
anomalies = iso_forest.fit_predict(data_scaled)
data_pivot['anomaly'] = anomalies

# Visualize anomalies
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_pivot['Social Media'], y=data_pivot['Productivity'], hue=data_pivot['anomaly'], palette='coolwarm')
plt.title('Anomaly Detection in App Usage')
plt.xlabel('Social Media Usage (minutes)')
plt.ylabel('Productivity Usage (minutes)')
plt.legend(title='Anomaly')
plt.show()

# Recommendations based on clustering
def generate_recommendations(cluster):
    recommendations = {
        0: 'Consider reducing Social Media usage to improve productivity.',
        1: 'You are in a balanced state. Maintain your current usage habits.',
        2: 'Focus on increasing time spent on educational apps to balance your usage.'
    }
    return recommendations.get(cluster, 'No recommendations available.')

data_pivot['recommendation'] = data_pivot['cluster'].apply(generate_recommendations)

# Display recommendations
print(data_pivot[['Social Media', 'Productivity', 'recommendation']].head())

# Save the results to a CSV file for presentation
data_pivot.to_csv('/content/digital_wellbeing_tracker_results.csv', index=False)
```
